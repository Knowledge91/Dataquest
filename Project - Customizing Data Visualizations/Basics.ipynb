# 3: Scatter Matrix, Part 1

```python
# Scatter plot that Pandas generates
from pandas.tools.plotting import scatter_matrix
scatter_matrix(recent_grads[["ShareWomen","Unemployment_rate"]])
```

```python
fig = plt.figure(figsize=(9,9))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.hist(new_recent_grads['ShareWomen'])
ax2.scatter(new_recent_grads['Unemployment_rate'],new_recent_grads['ShareWomen'])
ax3.scatter(new_recent_grads['ShareWomen'], new_recent_grads['Unemployment_rate'])
ax4.hist(new_recent_grads['Unemployment_rate'])
```

# 4: Scatter Matrix, Part 2

```python
fig = plt.figure(figsize=(9,9))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.hist(new_recent_grads['ShareWomen'])
ax2.scatter(new_recent_grads['Unemployment_rate'],new_recent_grads['ShareWomen'])
ax3.scatter(new_recent_grads['ShareWomen'], new_recent_grads['Unemployment_rate'])
ax4.hist(new_recent_grads['Unemployment_rate'])

ax1.xaxis.set_visible(False)
ax2.xaxis.set_visible(False)
ax2.yaxis.set_visible(False)
ax4.yaxis.set_visible(False)
ax1.set_ylabel("ShareWomen")
ax3.set_ylabel("Unemployment_rate")
ax3.set_xlabel("ShareWomen")
ax4.set_xlabel("Unemployment_rate")
fig.subplots_adjust(wspace=0, hspace=0)
```

# 5: Scatter Matrix, Part 3

```python
# First, we need to create a Figure instance and create 4 axes, corresponding to each subplot.
fig_rg = plt.figure(figsize=(9,9))
ax1_rg = fig_rg.add_subplot(2,2,1)
ax2_rg = fig_rg.add_subplot(2,2,2)
ax3_rg = fig_rg.add_subplot(2,2,3)
ax4_rg = fig_rg.add_subplot(2,2,4)

# Top left plot
ax1_rg.hist(new_recent_grads["ShareWomen"])
# Top right plot
ax2_rg.scatter(new_recent_grads["Unemployment_rate"], new_recent_grads["ShareWomen"])
# Bottom left plot
ax3_rg.scatter(new_recent_grads["ShareWomen"], new_recent_grads["Unemployment_rate"])
# Bottom right plot
ax4_rg.hist(new_recent_grads["Unemployment_rate"])

# Top left plot
ax1_rg.set_ylabel("ShareWomen")
ax1_rg.get_xaxis().tick_top()
ax1_rg.get_xaxis().set_visible(False)
ax1_rg.get_yaxis().tick_left()
ax1_rg.set_ylim(0,30)
ax1_rg.set_yticklabels([0, 5, 10, 15, 20, 25, 30])

# Top right plot
ax2_rg.get_yaxis().set_visible(False)
ax2_rg.get_xaxis().set_visible(False)
ax2_rg.set_xlim(0.0, 0.20)

# Bottom left plot
ax3_rg.set_xlabel("ShareWomen")
ax3_rg.set_ylabel("Unemployment_rate")
ax3_rg.get_xaxis().tick_bottom()
ax3_rg.get_yaxis().tick_left()
ax3_rg.set_ylim(0.00, 0.20)
ax3_rg.set_yticklabels([0.00, 0.05, 0.10, 0.15])
ax3_rg.set_xlim(0.0, 1.0)
ax3_rg.set_xticklabels([0.0, 0.2, 0.4, 0.6, 0.8], rotation=90)

# Bottom right plot
ax4_rg.set_xlabel("Unemployment_rate")
ax4_rg.get_yaxis().set_visible(False)
ax4_rg.get_xaxis().tick_bottom()
ax4_rg.set_xlim(0.0, 0.20)
ax4_rg.set_xticklabels([0.00, 0.05, 0.10, 0.15, 0.20], rotation=90)

# Remove space between sub plots.
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

# 6: Grouped Bar Plot, Part 1

```python
pd.options.mode.chained_assignment = None  # default='warn'
arts = new_recent_grads[new_recent_grads["Major_category"] == "Arts"]
arts["ShareMen"] = 1-arts["ShareWomen"]
arts.set_index("Major", inplace=True)
arts[["ShareMen", "ShareWomen"]].plot(kind="bar")
plt.show()
```

```python
new_recent_grads["ShareMen"] = new_recent_grads["Men"] / new_recent_grads["Total"]
arts = new_recent_grads[new_recent_grads["Major_category"] == "Arts"]
fig = plt.figure(figsize=(9,9))
ax1 = fig.add_subplot(1,1,1)
import numpy as np
locs = np.arange(arts.shape[0])
a = 0.4
bar_1 = ax1.bar(locs, arts["ShareMen"].tolist(), a)
ax1.set_xticklabels(arts['Major'].tolist(), rotation=90)
offset_locs = locs + 0.35
bar_2 = ax1.bar(offset_locs, arts["ShareWomen"].tolist(), a, color="green")
ax1.set_xticks(offset_locs)

plt.legend((bar_1, bar_2), ("ShareMen", "ShareWomen"), loc="upper left")
plt.grid()
plt.show()
```
