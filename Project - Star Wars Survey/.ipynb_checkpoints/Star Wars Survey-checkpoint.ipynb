## 2. Introduction
---

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

## 3: Cleaning Up Yes/No Columns
---

```python
# print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts())
# star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()

yes_no = {
    "Yes": True,
    "No": False
}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
star_wars
```

## 4: Cleaning Up Checkbox Columns
---

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
     np.nan: False,
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)

star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6",
})    
    
#star_wars.head()
```

## 5: Cleaning Up Ranking Columns
---

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6",
}) 
star_wars.head()
```

## 6: Finding The Highest Ranked Movie
---

```python
print(star_wars[star_wars.columns[9:15]].mean())

%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(range(6),star_wars[star_wars.columns[9:15]].mean())
```

### Review
The people liked the orignial 3 movies more than the newer ones. Movie 5 was liked most.

## 7: Finding The Most Seen Movie
---

```python
star_wars.columns
star_wars[star_wars.columns[3:9]].sum()
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

### View counts
Here we plotted the amount of seen movies (e.g. the first episode has been seen a little bit more than 650 times). As aspected the original movies have been seen more often than the newer ones. But the 4th (also original) has been seen less then episode 4 and 5 and the first movie plays in the same league as the original ones (concerining seen movies).

## 8: Exploring The Data By Binary Segments
---

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

fig = plt.figure(figsize=(20,12))

ax1 = fig.add_subplot(2,2,1)
ax1.bar(range(6),males[males.columns[3:9]].mean())
ax1.set_title('Males View Count')

ax2 = fig.add_subplot(2,2,2)
ax2.bar(range(6), females[females.columns[3:9]].sum())
ax2.set_title('Females View Count')

ax3 = fig.add_subplot(2,2,3)
ax3.bar(range(6),males[males.columns[9:15]].mean())
ax3.set_title('Males Rating Count')

ax4 = fig.add_subplot(2,2,4)
ax4.bar(range(6), females[females.columns[9:15]].sum())
ax4.set_title('Females Rating Count')


plt.show()
```

### Male vs Female
Interestingly, more males watches episodes 1-3, but males liked them far less than females did.
